```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model, preprocessing
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("car.data")
data.head()
```

```python
le = preprocessing.LabelEncoder()

buying = le.fit_transform(data["buying"])
maint = le.fit_transform(data["maint"])
door = le.fit_transform(data["door"])
persons = le.fit_transform(data["persons"])
lug_boot = le.fit_transform(data["lug_boot"])
safety = le.fit_transform(data["safety"])
classe = le.fit_transform(data["class"])

print(buying)
```

```python
x = list(zip(buying, maint, door, lug_boot, safety))
y = classe
print(x, y)

x_train, x_test, y_train, y_test  = sklearn.model_selection.train_test_split(x, y, test_size = 0.1)
```

```python
best= 0
for i in range(1, 10):
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(x_train, y_train)
    acc = model.score(x_test, y_test)
    print(i, acc)
    if(acc > best):
        best = acc
        j = i
        
model = KNeighborsClassifier(n_neighbors = j)
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print("best with", x, "neighbors and ",acc ,"accuracy")
```

```python
# testing model
names = ["unacc", "acc", "good", "vgood"]
predicted = model.predict(x_test)
for i in range(len(predicted)):
    print("pretected:", names[predicted[i]], ", actual data:", names[y_test[i]], ", data:",x_test[i])
```
